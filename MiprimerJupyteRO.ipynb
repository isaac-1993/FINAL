# Primer ejercicio, crear una distribución Normal de 30 números aleatorios

```r
x<-rnorm(30)
```

```r
x
```

## Ahora es posible crear un histograma de la distribución 

```r
hist(x)
```

## Es posible modificar los distintos elementos del histograma (color, límites y etiqueta de ambos ejes, etc.)

```r
hist (x, main="Histograma primer ejercicio", col="pink", border="yellow", labels=T, xlab="X", ylab="Frecuencia", ylim=c(0,10))
```
